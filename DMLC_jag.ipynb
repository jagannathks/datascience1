```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
train = pd.read_csv("./input/DMLC_data_Jag.csv")
print("train : " + str(train.shape))
```

```python
train.head()
```

```python
plt.scatter(train.AgeRange, train.ReflexTime, c = "blue", marker = ".", s=3000)
plt.title("Visualization - Age, ReflexTime")
plt.xlabel("AgeRange")
plt.ylabel("ReflexTime")
plt.ylim([0, 2])
plt.yticks(np.arange(0, 1.8, 0.3))
#plt.rcParams['figure.figsize'] = [10, 8]
plt.show()
```

```python
plt.scatter(train.NativeEnglish, train.WordsRecall, c = "blue", marker = ".")
plt.title("Visualization - Age, ReflexTime")
plt.xlabel("NativeEnglishSpeaker")
plt.ylabel("Words Recalled")
plt.ylim([0, 20])
plt.yticks(np.arange(0, 20, 1))
plt.show()
```

```python
plt.scatter(train.ReflexTime, train.WordsRecall, c = "blue", marker = ".", s=200)
plt.title("Visualization - ReflexTime, WordsRecall")
plt.xlabel("ReflexTime")
plt.ylabel("Words Recalled")
plt.xlim([0, 3])
plt.ylim([0, 18])
plt.xticks(np.arange(0, 2, 0.1))
plt.yticks(np.arange(0, 20, 1))
plt.show()
```

```python
plt.boxplot(train.AgeRange, train.ReflexTime)
plt.title("Visualization - Age, ReflexTime")
plt.xlabel("AgeRange")
plt.ylabel("ReflexTime")
plt.ylim([0, 2])
plt.yticks(np.arange(0, 1.8, 0.3))
plt.rcParams['figure.figsize'] = [40, 25]
plt.show()
```
